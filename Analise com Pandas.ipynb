# Descobrindo lucro por ação
 * Para descobrir o lucro por ação é necessario fazer o calculo do lucro liquido dividido pela quantidade de ações 

```python
dados_pl['lucro_por_acao'] = dados_pl['lucro_liquido']/dados_pl['quant_acoes']
dados_pl
```

# Descobrindo Preço sobre o Lucro
 * Para descobrir Preço sobre o lucro precisa calcular o preço dividido pelo lucro 

```python
dados_pl['preco_lucro'] = dados_pl['preco']/dados_pl['lucro_por_acao']
dados_pl
```

# Empresas que tiveram Preço sobre o Lucro positivo

```python
empresas_lucrativas = dados_pl[dados_pl['preco_lucro'] > 0 ]
empresas_lucrativas
```
